# Ejercicio para convertir temperaturas Celsius a Fahrenheit

```python
celsius = float(input('Dime la temperatura en ºC: '))
fahrenheit = celsius * (9/5) + 32
print('La temperatura de', celsius,'ºC = ',fahrenheit,'ºF')
kelvin = (fahrenheit - 32)/1.8 + 273.15
print('La temperatura es de', fahrenheit,'ºF = ',kelvin,'ºK')
```
