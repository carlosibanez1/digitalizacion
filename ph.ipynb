```python
ph=int(input("cuanto ph tiene:"))
if 0 < ph <= 4 :
  print("Es muy acido")
elif 4 < ph < 7:
  print("Es poco acido")
elif ph == 7 :
  print("Es neutro")
elif 7 < ph <= 10:
  print("Es poco alcalino")
elif 10 < ph <= 14:
  print("Es muy alcalino")
```
