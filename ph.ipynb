```python
ph=int(input("cuanto ph tiene:"))
if ph <= 4 :
  print("Es muy acido")
elif ph < 7 and ph > 4:
  print("Es poco acido")
elif ph == 7 :
  print("Es neutro")
elif ph <= 10 and ph > 7 :
  print("Es poco alcalino")
elif ph <= 14 and ph > 10 :
  print("Es muy alcalino")
```
